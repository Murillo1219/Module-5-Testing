When running the tests, some assertions raised errors. For example, the test_sum_tuple() function expected the sum of (1, 2, 2) to be 6, but the actual sum is 5. This caused Python to raise an AssertionError, showing that the test failed and indicating which function did not pass.

This demonstrates how unit testing works: it automatically checks whether the code produces the expected results. A failing test signals that the function does not behave as expected and needs to be corrected. Successful tests confirm correct behavior, while failed tests help identify and fix problems in the code.
